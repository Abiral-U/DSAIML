```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp, ttest_ind, f_oneway
```

image.png

```python
# 1. One-sample t-test
acorn_sample = [
    18.8, 6.6, 9.5, 11.2, 10.2, 7.4, 8.0, 9.6, 9.9, 9.0,
    7.6, 7.4, 10.4, 11.1, 8.5, 10.0, 11.6, 10.7, 10.3, 7.0
]
population_mean = 10
alpha = 0.05
```

```python
# Perform one-sample t-test
t_stat_1, p_value_1 = ttest_1samp(acorn_sample, population_mean)
```

```python
print("1. One-sample t-test:")
print(f"T-statistic: {t_stat_1:.3f}, P-value: {p_value_1:.3f}")
if p_value_1 < alpha:
    print("Reject the null hypothesis: The average mass of all acorns is different from 10 g.")
else:
    print("Fail to reject the null hypothesis: No sufficient evidence to say the average mass is different from 10 g.")
```

```python
# 2. Independent (unpaired) two-sample t-test
up_wind_sample = [
    10.8, 10.0, 8.2, 9.9, 11.6, 10.1, 11.3, 10.3, 10.7, 9.7,
    7.8, 9.6, 9.7, 11.6, 10.3, 9.8, 12.3, 11.0, 10.4, 10.4
]
down_wind_sample = [
    7.8, 7.5, 9.5, 11.7, 8.1, 8.8, 8.8, 7.7, 9.7, 7.0, 9.0,
    9.7, 11.3, 8.7, 8.8, 10.9, 10.3, 9.6, 8.4, 6.6, 7.2, 7.6,
    11.5, 6.6, 8.6, 10.5, 8.4, 8.5, 10.2, 9.2
]
```

```python
# Perform independent two-sample t-test
t_stat_2, p_value_2 = ttest_ind(up_wind_sample, down_wind_sample, equal_var=True)
```

```python
print("\n2. Independent two-sample t-test:")
print(f"T-statistic: {t_stat_2:.3f}, P-value: {p_value_2:.3f}")
if p_value_2 < alpha:
    print("Reject the null hypothesis: The mass of acorns from down-wind trees is different from up-wind trees.")
else:
    print("Fail to reject the null hypothesis: No sufficient evidence to say the masses are different.")
```

```python
# 3. ANOVA test
section_A = [51, 45, 33, 45, 67]
section_B = [23, 43, 23, 43, 45]
section_C = [56, 76, 74, 87, 56]
```

```python
# Perform ANOVA test
f_stat, p_value_3 = f_oneway(section_A, section_B, section_C)
```

```python
print("\n3. ANOVA test:")
print(f"F-statistic: {f_stat:.3f}, P-value: {p_value_3:.3f}")
if p_value_3 < alpha:
    print("Reject the null hypothesis: The mean marks of students in the three sections are different.")
else:
    print("Fail to reject the null hypothesis: No sufficient evidence to say the mean marks are different.")
```
